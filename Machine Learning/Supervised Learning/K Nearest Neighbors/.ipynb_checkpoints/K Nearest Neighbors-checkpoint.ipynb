# I. Algorithm

## 1. Mathematics

### Norms:
Calculate distance between points.
$$||\mathbf{x}||_p = (|x_1|^p + |x_2|^p + \dots |x_n|^p)^{\frac{1}{p}}$$

Note:

__Manhattan distance__ ($p = 1$):
$$||\mathbf{z} - \mathbf{x_i}||_1 = |\mathbf{z} - \mathbf{x_i}|(1)$$
__Euclidean distance__ ($p = 2$):
$$||\mathbf{z} - \mathbf{x_i}||_2^2 = (\mathbf{z} - \mathbf{x_i})^T(\mathbf{z} - \mathbf{x_i}) = ||\mathbf{z}||_2^2 + ||\mathbf{x_i}||_2^2 - 2||\mathbf{x_i}||^T||\mathbf{z}||(2)$$
- The formula is squared to avoid calculate square root.  

### Weight:
The way model treat the neighbors:
- "uniform" : each neighbor is treated equally.
- "distance" : nearer neigbor has more weight - the weight of each label is calculated by $\frac{1}{d}$. 

## 2. Code:
- Step 1: Use $(1)$ and $(2)$ to calculate distances.
- Step 2: Get label of the neighbors.
- Step 3: Label based on the weights of n nearest neigbors.

Note:
- np.choose is rewrite because of the limitation of this command.

```python
import numpy as np

# Rewrite np.choose
def choose(a,c):
    return np.array([c[a[I]] for I in np.lib.index_tricks.ndindex(a.shape)])

class KNearestNeighbors(object):
    def __init__(self, n_neighbors, weight = "uniform", p = 2):
        self.X_train = np.array([])
        self.y_train = np.array([])
        self.X_test = np.array([])
        self.y_pre = np.array([])
        self.distance = np.array([])
        self.n_neighbors = n_neighbors
        self.weight = weight
        self.p = p

    def fit(self,X_train,y_train):
        self.X_train = X_train
        self.y_train = y_train
        
    # Step 1: Calculate theta
    def calculate_distance(self):
        if self.p == 1:           
            self.distance = np.abs(X_test[:,None] - X_train).sum(-1) 
        if self.p == 2:
            sum_X2_train = np.sum(self.X_train*self.X_train, 1).reshape(1, -1)
            sum_X2_test = np.sum(self.X_test*self.X_test, 1).reshape(-1, 1)
            self.distance = sum_X2_train + sum_X2_test - 2*self.X_test.dot(self.X_train.T)
            
    # Step 2: Get label of the neighbor
    def get_labels(self):
        arg_distance = np.argsort(self.distance,axis = 1)[:, :self.n_neighbors]
        label = choose(arg_distance, self.y_train).reshape(-1,self.n_neighbors)
        return label
    
    # Step 3: Label based on the weights of n nearest neigbors
    def predict(self, X_test):
        self.X_test = X_test
        self.calculate_distance()
        labels = self.get_labels()
        uniques = np.unique(labels)
        result = np.zeros((labels.shape[0],uniques.shape[0]))
        if self.weight == "uniform":
            for i in range(uniques.shape[0]):
                zeros = np.zeros((labels.shape[0],labels.shape[1]))
                zeros[labels == i] = 1
                result[:,i] = zeros.sum(axis = 1)
                self.y_pre = result.argmax(axis = 1)
            return self.y_pre
        if self.weight == "distance":
            for i in range(uniques.shape[0]):
                zeros = np.zeros((labels.shape[0],labels.shape[1]))
                zeros[labels == i] = 1
                neighbors_distance = np.sort(self.distance,axis = 1)[:, :self.n_neighbors]
                if self.p == 2:
                    neighbors_distance = np.sqrt(neighbors_distance)
                neighbors_distance[ neighbors_distance == 0] = 0.00001
                result[:,i] = (zeros/neighbors_distance).sum(axis = 1)
                self.y_pre = result.argmax(axis = 1)
            return self.y_pre
```

# II. Practice

Apply K nearest neighbors on Iris data set
![](iris.png)

```python
# Load data
from sklearn import neighbors, datasets
from sklearn.model_selection import train_test_split
np.random.seed(7)
iris = datasets.load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=130)

# Plot data
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
label_mapping = dict(list(enumerate(iris.target_names)))
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
    # Plot 1
plot_1 = sns.scatterplot(x = X_train[:,0], y=X_train[:,1], hue = pd.Series(y_train).map(label_mapping), ax=axes[0])
plot_1.set_xlabel(iris.feature_names[0].capitalize())
plot_1.set_ylabel(iris.feature_names[1].capitalize())
    # Plot 2
plot_2 = sns.scatterplot(x = X_train[:,2], y=X_train[:,3], hue = pd.Series(y_train).map(label_mapping), ax=axes[1])
plot_2.set_xlabel(iris.feature_names[2].capitalize())
plot_2.set_ylabel(iris.feature_names[3].capitalize())
plt.show()
```

```python
# Create model
model = KNearestNeighbors(3, p = 1, weight = "distance")
model.fit(X_train,y_train)

# Predict 
y_pre = model.predict(X_test)

# Plot model
label_mapping = dict(list(enumerate(iris.target_names)))
dataset_mapping = dict(list(enumerate(["training set","test set"])))
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 5))
plot_train = np.concatenate((X_train,y_train.reshape(-1,1), np.zeros((y_train.shape[0],1))), axis = 1)
plot_pre = np.concatenate((X_test,y_pre.reshape(-1,1), np.ones((y_pre.shape[0],1))), axis = 1)
plot = np.concatenate((plot_train,plot_pre), axis = 0)
    # Plot 1
plot_1 = sns.scatterplot(x = plot[:,0], y= plot[:,1], hue = pd.Series(plot[:,4]).map(label_mapping), style = pd.Series(plot[:,5]).map(dataset_mapping), ax=axes[0])
plot_1.set_xlabel(iris.feature_names[0].capitalize())
plot_1.set_ylabel(iris.feature_names[1].capitalize())
    # Plot 2
plot_2 = sns.scatterplot(x = plot[:,2], y= plot[:,3], hue = pd.Series(plot[:,4]).map(label_mapping), style = pd.Series(plot[:,5]).map(dataset_mapping), ax=axes[1])
plot_2.set_xlabel(iris.feature_names[2].capitalize())
plot_2.set_ylabel(iris.feature_names[3].capitalize())
plt.show()

# Accuracy
from sklearn.metrics import accuracy_score
print(100*accuracy_score(y_test, y_pre))
```
